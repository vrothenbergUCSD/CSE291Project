```python
!pip install openai
import time
import openai
import pandas as pd
from tqdm import tqdm
import os
from openai import OpenAI

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/summarized_data1.csv", on_bad_lines = 'skip').reset_index()
```

```python
def summarize_report(r1, r2):
  query = "You're a medical expert; here are summaries for one patient: {}\n{}\n Can you determine the cause of readmission. Be very concise, try to be as technical as possible, use one sentence:".format(r1, r2)
  client = OpenAI(api_key="")
  response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You summarize any text that is given to you."},
        {"role": "user", "content": query}
        ]
    )

  time.sleep(41)
  return response.choices[0].message.content
```

```python
openai.api_key = ""
tqdm.pandas()
df["CAUSES"] = ""
df["CAUSES"] = df.progress_apply(lambda i: summarize_report(i.Report1, i.Report2), axis = 1)
#t1 = """A 69-year-old male recently had a CABG. He felt a "click" when he coughed. He was short of breath with increasing fatigue and unable to do his daily chores. He had no known allergies to drugs."""
#t2 = """69-year-old male with recent history of CABG presented with sternal debridement and unstable stenotomy. New large presternal fluid collection is present, greatest in space between the topmost and secondmost plates."""
#summary = summarize_report(t1, t2)
```
